```python
# First do the necessary imports
from __future__ import print_function, division
%matplotlib inline

import thinkdsp
import thinkplot
import numpy as np
from IPython.display import Audio
```

```python
# now load the infamous recording
wave = thinkdsp.read_wave('yORl.wav')
```

```python
# listen to it
wave.make_audio()
```

```python
# pick the part where it obviously says 'Yanny'
start = 1.1
duration = 0.9
segment = wave.segment(start, duration)
segment.make_audio()
```

```python
# plot this segment
segment.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
# create the freq spectrum
spectrum = segment.make_spectrum()
spectrum.plot(high=12000)
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
# create a low pass filtered version
#spectrum.low_pass(1000)

# if you are a Laurel person you can try applying a high pass filter
spectrum.high_pass(1000)

spectrum.plot()
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
filtered = spectrum.make_wave()
filtered.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
segment.plot()
thinkplot.config(xlabel='Time (s)')
```
